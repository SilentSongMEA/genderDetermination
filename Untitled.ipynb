```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
import numpy as np
```

```python
model=YOLO("yolov8n.pt")
```

```python
video_path=r"C:\Users\Abouzarrayaneh\Desktop\genderDetermination\easyGerman.mp4"
# cap=cv2.VideoCapture(0)
cap=cv2.VideoCapture(video_path)

ret,frame=cap.read()
cap.release()

frame_rgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

results=model(frame_rgb)

annotated_frame=results[0].plot()

plt.figure(figsize=(10,6))
plt.imshow(annotated_frame)
plt.axis('off')
plt.title('gender determination with yoloV8')
plt.show()
```

```python
results = model(frame_rgb) 
for box in results[0].boxes:
    cls = int(box.cls[0])
    conf = float(box.conf[0])
    
    if cls != 0 or conf < 0.5:
        continue

    x1, y1, x2, y2 = map(int, box.xyxy[0])
    width = x2 - x1
    height = y2 - y1

    if width < 50 or height < 50:
        continue

    face_crop = frame_rgb[y1:y2, x1:x2]

    try:
        analysis = DeepFace.analyze(face_crop, actions=["gender"], enforce_detection=True)
        gender = analysis[0]['gender']
        print(f"gender detection: {gender}")
        
        gender_label = max(gender, key=gender.get)
        
        cv2.rectangle(frame_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame_rgb, gender_label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
    except Exception as e:
        print("Error gender detection", e)

plt.figure(figsize=(10,6))
plt.imshow(frame_rgb)
plt.axis('off')
plt.title("gender in video and picture")
plt.show()
        
```

```python
video_path=r"C:\Users\Abouzarrayaneh\Desktop\genderDetermination\easyGerman.mp4"
# cap=cv2.VideoCapture(0)
cap=cv2.VideoCapture(video_path)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = model(frame_rgb)

    male_count = 0
    female_count = 0

    for box in results[0].boxes:
        cls = int(box.cls[0])
        conf = float(box.conf[0])

        if cls != 0 or conf < 0.5:
            continue

        x1, y1, x2, y2 = map(int, box.xyxy[0])
        width = x2 - x1
        height = y2 - y1

        if width < 50 or height < 50:
            continue

        face_crop = frame_rgb[y1:y2, x1:x2]

        try:
            analysis = DeepFace.analyze(face_crop, actions=["gender"], enforce_detection=True)
            gender = analysis[0]['gender']
            gender_label = max(gender,
                               key=gender.get)

            if gender_label == "Man":
                male_count += 1
            else:
                female_count += 1

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, gender_label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
        except Exception as e:
            print("Error gender detection:", e)

    # نمایش تعداد مرد و زن
    cv2.putText(frame, f"Males: {male_count}  Females: {female_count}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    cv2.imshow("Gender Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print(f"count Male={male_count} | count Female={female_count}")
```
